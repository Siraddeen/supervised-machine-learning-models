# Supervised Machine Learning With train-test-split():  Linear Regression

```python
# let us import required pakages

import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

```python
 # let us create new data variables here
    
x = np.arange(20).reshape(-1,1)         # having one column
y = np.array([5,12,11,19,30,29,23,40,51,54,74,62,68,73,89,84,89,101,99,106])
```

```python
x
```

```python
y
```

```python
# let us create train data & test data from the        created data set

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 8 , random_state = 1)

                          # here i created (or) wanted only raw 8 value in number for test size
```

```python
x_train
```

```python
x_test
```

### let us build linear regression model 

```python
model = LinearRegression()    # here you can give any name in place of model 
```

```python
# now  let us make some training to our model

model. fit(x_train,  y_train  )
                                    # now LR model is created for us
```

```python
# you know that formula of LR is     y = mx+c
# so let us find    m & c   with for c use .intercept_         &   for  m   use   .coef_

model.intercept_
                              # c = 5.32
```

```python
model.coef_            #  m = 5.24
```

```python
# from general equation  y = mx +c     
        #                y = 5.24 x   + 5.32             we will find every y value 
    
```

```python
# let us see some pridicted values sos you can understand better

model.predict(x_test)    # if we put x value inside () it will give us  result in y value 
```

```python
# don't worry if you not understanding them yet i will show them all side by side

# let's check our model score i.e how good it in predicting of y

model.score(x_train,y_train)       # seems good result
```

```python
model.score(x_test,y_test)         # seems ok result
```

```python
# remember  if testing data have higher score than training data  then your model did not learn well
```

```python
# now let me create dataframes of obtained values & show them side by side

model_prediction = model.predict(x_test) 
```

```python
model_prediction         # now convert it into dataframe
```

```python
Y_pred = pd.DataFrame(model_prediction,columns=['Y'])
```

```python
# now let me put x in place of x_test
model_prediction = model.predict(x)
Y_pred = pd.DataFrame(model_prediction,columns=['Y_predict'])
x =pd.DataFrame(x,columns=['x'])
y =pd.DataFrame(y,columns=['y'])
fram = [x,y]
pd.concat(fram)
df = pd.concat(fram,axis=1)
y_frames = [df, Y_pred]

df1 = pd.concat(y_frames,axis=1)
```

```python
  # if x=4 & y predict is   =        5.24*4 + 5.32     ==  26.28         check above
```

```python
# now let us check for errors        Mean Absolute Error      & root mean square error

from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(df.y,df1.Y_predict)    
                 # here   y obtained from df dataframe      & y_predict obtained  from df1 dataframe
```

```python
MAE
```

```python
# for RMSE   use
RMSE = np.sqrt(MAE)
```
